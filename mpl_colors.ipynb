# Colores de matplotlib en el módulo mpl_colors

Matplotlib permite representar cualquier color a través de su código de color (https://matplotlib.org/stable/tutorials/colors/colors.html)

Matplotlib permite trabajar con nombres de colores. Los colores se agrupan en tablas de colores con nombre y en mapas de colores (colormaps). Los colors maps proporcionan generalente transiciones suaves entre varios colores. Los colores en cada colormap se diferencian por su código de color, pero no tienen nombre.

# Colormaps

Cada colormap está formado por un conjunto diferentes de colores que determina la suavidad del cambio entre colores sucesivos y su discriminación visual. La función __print_colormaps_names__ nos muestra los nombres de los colormaps disponibles y el número de colores en cada colormap.

```python
mplc.print_colormaps_names()
```

## Ver los colores en cada colormap

Los nombres de los colormaps nos dan una indicación de su contenido, pero lo que nos interesa es ver esos colores. Para esto se utiliza la función __display_colormaps__. Como parámetro es el listado de los colormaps que quieres visualizar. Si no le pasas ningún parámetro, display_colormaps(), se muestran todos los colormaps.

```python
mplc.display_colormaps(['tab10', 'tab20', 'brg', 'hsv', 'gist_rainbow', 'twilight_shifted'])
```

## Extraer colores de un colormap

En la celda previa vemos todos los colores en cada colormap. Ahora vamos a ver los colores que extraemos de cada colormap cuando hacemos una petición de un número determinado de colores, para lo que se utiliza la función __display_colors_in_colormap__.

La función __display_colors_in_colormap__ tiene tres parámetros, cada uno con un valor por defecto: 

* El nombre del colormap (por defecto 'hsv')
* El número de colores que se van a mostrar (por defecto 10).
* Si los colores se van a extraer correlativos desde el principio de la lista de colores o se extraerán buscando la máxima separación entre los colores: por ejemplo, si pedimos 3 colores se mostraán el primero, el medio y el útimo)

Si el número de colores que se desean ver excede el de colores en el colormap, habrá puntos con el mismo color. 

Haz una prueba utilizando:

* El colormap 'tab10' con 10 puntos y con 20 puntos; verás que con 20 puntos cada dos puntos tienen el mismo color.
* El colormap 'tab10' con 3 puntos y asignado al parámetro correlative los valores Tue y False

```python
mplc.display_colors_in_colormap('tab10', 3, correlative=True)
```

# Tablas de colores con nombre

Se puede asignar colores mediante nombres en vez de por su código de color. Para ello matplotlib suministra los nombres de unas tablas de colores que a su vez contienen unos colores a los que te puedes referir por sus nombres. Las listas de colores disponibles son:

* CSS4_COLORS : Contiene 147 colores con nombre definidos en la especificación CSS4.
* CSS4_COLORS_EXTENDED : Contiene los colores de la tabla CSS4_COLORS más 102 colores adicionales.
* HTML4_COLORS : Contiene 48 colores con nombre definidos en la especificación HTML4.
* TABLEAU_COLORS : Contiene 10 colores utilizados en la tabla de color Tableau.
* XKCD_COLORS : Contiene 312 colores con nombe de la encuesta de color XKCD.

```python
# Sample dictionary d1
d1 = {
    "apple": "red",
    "banana": "yellow",
    "cherry": "red",
    "date": "brown"
}

# Swap keys and values, and create a new dictionary d2
d2 = {v: k for k, v in d1.items()}

# Sort d2 by its keys
sorted_d2 = dict(sorted(d2.items()))

print(sorted_d2)
```

```python
d1 = {"apple": 3, "banana": 1, "cherry": 2}

# Sort the dictionary by values and create a list of sorted key-value pairs
sorted_items = sorted(d1.items(), key=lambda x: x[1])

# Create a new dictionary from the sorted list
sorted_dict = dict(sorted_items)

print(sorted_dict)
```
