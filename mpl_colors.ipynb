# Colores de matplotlib en el módulo mpl_colors

Matplotlib permite representar cualquier color a través de su código de color (https://matplotlib.org/stable/tutorials/colors/colors.html)

Matplotlib permite trabajar con nombres de colores. Los colores se agrupan en tablas de colores con nombre y en mapas de colores (colormaps). Los colors maps proporcionan generalente transiciones suaves entre varios colores. Los colores en cada colormap se diferencian por su código de color, pero no tienen nombre.

```python
import matplotlib.colors as mcolors

from mpl_colors import MplColors as mplc
```

# Nombres de colores

Matplotlib tiene un diccionario con los nombres de colores asociado a un código de color hexadecimal RGB:

```python
print('Número de colores con nombre en matplotlib:', len(mcolors.cnames))
print('Nombres de colores en matplotlib:\n', mcolors.cnames.keys())
```

```python
# Colors displayed by color name
mplc.display_all_named_colors()
```

```python
# Colors displayed by hex RGB code
mplc.display_all_named_colors(sorted_by_name=False)
```

Para visualizar algunos de los colores con nombre utilizamos la función __display_some_named_colors__. Como parámetro tiene una lista con los colores que queremos visualizar; por defecto se muestran los 3 colores del modelo RGB. 

```python
mplc.display_some_named_colors()
```

```python
mplc.display_some_named_colors(['aqua', 'aquamarine', 'azure'])
```

# Tablas de colores con nombres
Matplotlib ofrece diccionarios de colores con nombre asociado a su código de color.

Las tablas de colores disponibles se muestran con la función __get_color_table_names()__:

```python
mplc.get_color_table_names()
```

Para ver todos los colores en cada color table utilizamos la función __display_color_tables()__:

```python
mplc.display_color_tables()
```

Para ver algunos mejor algunos colores y sus nombres en las color tables de muchos colores podemos utilizar la función __display_colors_in_color_table__, que tiene dos parámetros

* ctable_name, nombre de la color table (por defecto 'tableau')
* ncolors, número de colores que se muestran (10 por defecto).

Para un número de colores dados, los colores son seleccionados para que haya la mayor separación entre ellos en la lista de la tabla de colores.

```python
mplc.display_colors_in_color_table(ctable_name='css4', ncolors=15)
```

# Colormaps

A diferencia de las tablas de colores, los mapas de colores (colormaps) son conjuntos de colores sin nombre, lo que permite crear tablas de muchos colores y tonalidades. La mayor parte están diseñadas para conseguir conjuntos de colores con cambios suaves entre ellos. Otras reproducen las tablas de colores. La función __print_colormaps_names__ nos muestra los nombres de los colormaps disponibles y el número de colores en cada colormap.

```python
mplc.print_colormaps_names()
```

## Ver los colores en cada colormap

Los nombres de los colormaps nos dan una indicación de su contenido, pero lo que nos interesa es ver esos colores. Para esto se utiliza la función __display_colormaps__. Como parámetro es el listado de los colormaps que quieres visualizar. Si no le pasas ningún parámetro, display_colormaps(), se muestran todos los colormaps (el proceso se puede interrumpir a intervalor regulares).

```python
mplc.display_colormaps(['tab10', 'tab20', 'brg', 'hsv', 'gist_rainbow', 'twilight_shifted'])
```

## Extraer colores de un colormap

Algunos colormaps tienen muchos colores y queremos verlos con más detalle. Para realizar esta tarea tenemos dos funciones: __display_colors_in_colormap__ y __get_colors_in_colormap__.

La función __display_colors_in_colormap__ muestra un número de colores determinado seleccionados de manera que la separación entre colores sea máxima (por ejemplo, si en colormap de 256 colores pedimos 3 colores, devolvera el primero, el último y el del medio). Los parámetros de la esta función son: 

* El nombre del colormap (por defecto 'hsv')
* El número de colores que se van a mostrar (por defecto 10).

Si el número de colores que se desean ver excede el de colores en el colormap, habrá puntos con el mismo color. 

Haz una prueba utilizando:

* El colormap 'tab10' con 10 puntos y con 20 puntos; verás que con 20 puntos cada dos puntos tienen el mismo color.
* El colormap 'tab10' con 3 puntos y asignado al parámetro correlative los valores Tue y False

```python
mplc.display_colors_in_colormap('tab10', 10)
```

La otra función es __get_colors_in_colormap__. La función tiene 2 parámetros:

* cmap_name, nombre del colormap
* indexes, una lista de valores enteros que corrresponden a la posición de los colores que queremos extraer del colormap (recordemos que el colormap no tiene nombres de color); es posible extraer un rango consecutivo de índices con la función slice, por ejemplo slice(3:5) extrae los elementos 3, 4 y 5. Recuerda también que el primer elemento tiene índice 0.    

```python
cmap_name = 'gist_rainbow'
indexes = [slice(0,20)]

mplc.get_colors_in_colormap(cmap_name, indexes)
```

Podemos observar la gradualidad en los tonos de rojo, que son difícilmente distinguibles a la vista a pesar de corresponder a códigos de color diferentes

Prueba ahora con el colormap y los colores a extraer que quieras:

```python
cmap_name = 'hsv'
indexes = [slice(20,40)]

mplc.get_colors_in_colormap(cmap_name, indexes)
```
